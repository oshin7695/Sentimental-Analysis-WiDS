```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv("training.csv")
test_df=pd.read_csv("test.csv")
```

```python
df=df1.drop_duplicates(subset ="text", keep = False, inplace = False)
df['text'].value_counts()
```

```python
import re
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
corpus = []
for i in df.index:
    text= re.sub('[^a-z!@?$]', ' ', df['text'][i])
    text = text.split()
    lemmatizer = WordNetLemmatizer()  
    all_stopwords = stopwords.words('english') 
    all_stopwords.remove('not')
    newStopWords = ['www','http','ive','im','pe','soo']
    all_stopwords.extend(newStopWords)
    text = [lemmatizer.lemmatize(word,'v') for word in text if not word in set(all_stopwords)]
    text = [lemmatizer.lemmatize(word,'n') for word in text if not word in set(all_stopwords)]
    text = ' '.join(text)
    corpus.append(text)
print(corpus)  
```

```python
import re
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
test_corpus = []
for i in test_df.index:
    text1 = re.sub('[^a-z!@?$]', ' ', test_df['text'][i])
    text1 = text1.split()
    lemmatizer = WordNetLemmatizer()  
    all_stopwords = stopwords.words('english') 
    all_stopwords.remove('not')
    newStopWords = ['www','http','ive','im','pe','soo']
    all_stopwords.extend(newStopWords)
    text1 = [lemmatizer.lemmatize(word,'v') for word in text1 if not word in set(all_stopwords)]
    text1 = [lemmatizer.lemmatize(word,'n') for word in text1 if not word in set(all_stopwords)]
    text1 = ' '.join(text1)
    test_corpus.append(text1)
print(test_corpus)
```

```python
d = {}
i = 1
for j in range(len(corpus)):
    l = corpus[j].split()    
    for x in l:
        if x not in d:
             d[x]=i
             i += 1
```

```python
i = len(d)+1
m=1
for j in range(len(test_corpus)):
    l = test_corpus[j].split()    
    for x in l:
        if x not in d:
             d[x]=i
             i += 1
             
```

```python
len(d)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import losses
from tensorflow.keras import regularizers
from tensorflow.keras.constraints import unit_norm
```

```python
print(tf.__version__)
```

```python
vocab_size = len(d) 
model = keras.Sequential()
model.add(layers.Embedding(input_dim=len(d)+1, output_dim=200, mask_zero=True))
model.add(layers.Conv1D(32, 5, activation='relu'))
model.add(layers.GlobalMaxPooling1D())
#model.add(layers.Dropout(0.4)),
#model.add(keras.layers.Dense(units=32,  kernel_constraint=unit_norm(), activation='relu'))
# keras.layers.Dropout(rate=0.2), 
model.add(layers.Dropout(0.2)),
model.add(keras.layers.Dense(units=6, activation='softmax'))
```

```python
model.compile(optimizer=tf.optimizers.Adamax(0.005),loss='sparse_categorical_crossentropy',metrics=['acc'])
```

```python
model.summary()
```

```python
X_train=[]
for i in range(len(corpus)):
    a=[]
    l = corpus[i].split()
    for x in l:
        a.append(d[x])
    X_train.append(a)
X_train=tf.keras.preprocessing.sequence.pad_sequences(
    X_train, maxlen=None, dtype='int32', padding='post',
     value=0.0
)
```

```python
index_list = df.label.tolist()
```

```python
Y_train=[]
for i in range(len(index_list)):
    b=[]
    b.append(index_list[i])
    Y_train.append(b)
Y_train=tf.keras.preprocessing.sequence.pad_sequences(
    Y_train, maxlen=None, dtype='int32', padding='post',
    value=0.0
)
```

```python
#callback = tf.keras.callbacks.EarlyStopping(monitor='val_acc', patience=3)
model.fit(X_train, Y_train, verbose=1, validation_split=0.2, shuffle=True, epochs=7)
```

```python
X_test=[]
for i in range(len(test_corpus)):
    a=[]
    l = test_corpus[i].split()
    for x in l:
        a.append(d[x])
    X_test.append(a)
X_test=tf.keras.preprocessing.sequence.pad_sequences(
    X_test, maxlen=None, dtype='int32', padding='post',
     value=0.0
)
X_test
```

```python
y_pred= model.predict(X_test)
y_pred = np.argmax(y_pred, axis = 1)
print(y_pred)
```
