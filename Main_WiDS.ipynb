```python
frequency = {}

# iterating over the list
for item in df1['label']:
   # checking the element in dictionary
   if item in frequency:
      # incrementing the counr
      frequency[item] += 1
   else:
      # initializing the count
      frequency[item] = 1

# printing the frequency
print(frequency)
```

```python
df=df1.drop_duplicates(subset ="text", keep = False, inplace = False)
df['text'].value_counts()
```

```python
word_freq=df.text.str.split(expand=True).stack().value_counts()
word_freq={key:val for key, val in word_freq.items() if val > 50}
word_freq
```

```python
import re
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
corpus = []
for i in df.index:
    text= re.sub('[^a-z!@?$]', ' ', df['text'][i])
    text = text.split()
    lemmatizer = WordNetLemmatizer()  
    all_stopwords = stopwords.words('english') 
    newStopWords = ['www','http','ive','im','pe','soo']
    all_stopwords.extend(newStopWords)
    text = [lemmatizer.lemmatize(word,'v') for word in text if not word in set(all_stopwords)]
    text = ' '.join(text)
    corpus.append(text)
print(corpus)  
```

```python
import re
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
test_corpus = []
for i in test_df.index:
    text1 = re.sub('[^a-z!@?$]', ' ', test_df['text'][i])
    text1 = text1.split()
    lemmatizer = WordNetLemmatizer()  
    all_stopwords = stopwords.words('english') 
    newStopWords = ['www','http','ive','im','pe','soo']
    all_stopwords.extend(newStopWords)
    text1 = [lemmatizer.lemmatize(word,'v') for word in text1 if not word in set(all_stopwords)]
    text1 = ' '.join(text1)
    test_corpus.append(text1)
print(test_corpus)
```
